```python
import pandas as pd
```

```python
df = pd.read_csv('data/student_lifestyle_100k.csv')
```

```python
new_cols = [col.lower() for col in df.columns]
df.columns = new_cols
df.columns
```

```python
df.shape
```

```python
df = pd.get_dummies(df, drop_first=False)
df.columns
```

```python
#train test split
from sklearn.model_selection import train_test_split

X = df.drop(columns=['depression'])
y = df['depression']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
# train a naive bayes classifier
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
# create dummy variables for categorical features

model.fit(X_train, y_train)

# evaluate the model
from sklearn.metrics import classification_report, confusion_matrix
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# print confusion matrix
cm = confusion_matrix(y_test, y_pred)

tp ,fn, fp, tn = cm.ravel()

print(f'TP: {tp}, FN: {fn}, FP: {fp}, TN: {tn}')
```

```python
#down sampling
#upsampling
```

```python
df.depression
```

```python
#upsample sampling the majority class

df_majority = df[~df.depression]

df_majority.shape
```

```python
df_minority = df[df.depression]
df_minority.shape
```

```python
df_minority  = df_minority.sample(df_majority.shape[0], replace=True, random_state=42)
df_minority.shape
```

```python
print(df_majority.shape, df_minority.shape)
```

```python
new_df = pd.concat([df_majority, df_minority], axis=0)
new_df.depression.value_counts()
```

```python
# train a naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

X = new_df.drop(columns=['depression'])
y = new_df['depression']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = MultinomialNB()
# create dummy variables for categorical features

model.fit(X_train, y_train)

# evaluate the model
from sklearn.metrics import classification_report, confusion_matrix
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

# print confusion matrix
cm = confusion_matrix(y_test, y_pred)

tp ,fn, fp, tn = cm.ravel()

print(f'TP: {tp}, FN: {fn}, FP: {fp}, TN: {tn}')
```

```python
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)

model.fit(X_train, y_train)

# evaluate the model
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
# print confusion matrix
cm = confusion_matrix(y_test, y_pred)
tp ,fn, fp, tn = cm.ravel()
print(f'TP: {tp}, FN: {fn}, FP: {fp}, TN: {tn}')
```

```python
# random forest classifier

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)

# evaluate the model
from sklearn.metrics import classification_report, confusion_matrix
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

# print confusion matrix
cm = confusion_matrix(y_test, y_pred)
tp ,fn, fp, tn = cm.ravel()
print(f'TP: {tp}, FN: {fn}, FP: {fp}, TN: {tn}')
```
